```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
extortion = [1217, 1180, 1448, 1684]
x_axis = np.arange(len(extortion))
plt.bar(x_axis, extortion, color="grey", align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["2011", "2012", "2013", "2014"])
plt.xlabel("Year")
plt.ylabel("Number of Reported Incidents(Per Year)")
plt.title("Reported Extortion/Blackmail Incidents")
plt.savefig("../Project/Extortion_Blackmail.png")
```
