```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x_axis = [1978, 1982, 1986, 1990, 1992, 1994, 1996, 1999, 2004, 2004, 2008, 2012]
x_axis
```

```python
age_13_math = [264, 269, 269, 270, 273, 274, 274, 276, 281, 279, 281, 285]
age_9_math = [219, 219, 222, 230, 230, 231, 231, 232, 241, 239, 243, 244]
```

```python
age_13_math, = plt.plot(x_axis, age_13_math, marker="+",color="blue", linewidth=1, label="Avg Math Scores(Age 13)")
age_9_math, = plt.plot(x_axis, age_9_math, marker="s", color="Red", linewidth=1, label="Avg Math Scores (Age 9)")
```

```python
read_year = [1971, 1975, 1980, 1984, 1988, 1990, 1992, 1994, 1996, 1999, 2004, 2004, 2008, 2012]
age_13_read = [255, 256, 258, 257, 257, 257, 260, 258, 258, 259, 259, 257, 260, 263]
age_9_read = [208, 210, 215, 211, 212, 209, 211, 211, 212, 212, 219, 216, 220, 221]

age_13_read, = plt.plot(read_year, age_13_read, marker="+",color="black", linewidth=1, label="Reading Scores (Age 13)")
age_9_read, = plt.plot(read_year, age_9_read, marker="s", color="green", linewidth=1, label="Reading Scores (Age 9)")
```

```python
plt.xlabel("Year")
plt.ylabel("Average Test Scores")
plt.title("Math & Reading Test Scores")
plt.grid()
plt.legend(loc="best", frameon=False)
plt.savefig("../Project/Math_Reading_Test_Scores.png")
```
