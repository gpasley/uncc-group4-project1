```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fraud = [245301, 261718, 270106, 294160]
x_axis = np.arange(len(fraud))
plt.bar(x_axis, fraud, color="orange", align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["2011", "2012", "2013", "2014"])
plt.xlabel("Year")
plt.ylabel("Number of Reported Incidents(Per Year)")
plt.title("Reported Fraud Offenses")
plt.savefig("../Project/Fraud_Offenses.png")
```
