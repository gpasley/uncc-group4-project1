```python
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Loading the CSV
by_race_df = pd.read_csv("../Resources/By_race.csv.csv")
by_race_df.head()
```

```python
#Setting X Axis and Tick Locations
x_axis = np.arange(len(by_race_df))
tick_locations = [value for value in x_axis]
```

```python
#Creating a list indicating where to write X Labels
#White_People

plt.figure(figsize=(20,4))
plt.bar(x_axis, by_race_df["% of U.S. adults who use the internet, by race/ethncity"], color='r', alpha=1.0, align="center")
plt.xticks(tick_locations, by_race_df["Unnamed: 1"], rotation="vertical")
```

```python
#Creating a list indicating where to write X Labels
#Black_People

plt.figure(figsize=(20,4))
plt.bar(x_axis, by_race_df["% of U.S. adults who use the internet, by race/ethncity"], color='r', alpha=1.0, align="center")
plt.xticks(tick_locations, by_race_df["Unnamed: 2"], rotation="vertical")
```
