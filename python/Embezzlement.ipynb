```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
embezzlement = [17000, 17364, 18241, 18885]
x_axis = np.arange(len(embezzlement))
plt.bar(x_axis, embezzlement, color="green", align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["2011", "2012", "2013", "2014"])
plt.xlabel("Year")
plt.ylabel("Number of Reported Incidents(Per Year)")
plt.title("Reported Embezzlement Incidents")
plt.savefig("../Project/Embezzlement_Incidents.png")
```
