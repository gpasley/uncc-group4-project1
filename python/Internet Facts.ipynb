```python
# Dependencies
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import pandas as pd
```

```python
#------------------------------
#Social Media Usage by Age
df_age = pd.read_excel("Dataset/internet_social_media_facts/social_media_fact_sheet.xlsx", sheet_name=1, skiprows=1)
df_age = pd.DataFrame(df_age)
df_age = df_age.rename(columns={'Unnamed: 0':'date'})
df_age = df_age.set_index('date')
df_age.head()
```

```python
ax = df_age.plot(kind='line', color=['r','b','g','aqua'])
vals = ax.get_yticks()
ax.set_yticklabels(['{:.0%}'.format(x) for x in vals])
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.title("% of U.S. adults who uses social media, \n by age group")
plt.legend(loc='best')
plt.savefig('outputs/01_sm_use_total.png')
plt.show();
```

```python
#------------------------------
#Social Media Usage by Education Level
df_edu = pd.read_excel("Dataset/internet_social_media_facts/social_media_fact_sheet.xlsx", sheet_name=5, skiprows=1)
df_edu = pd.DataFrame(df_edu)
df_edu = df_edu.rename(columns = {'Unnamed: 0':'date'})
df_edu = df_edu.set_index('date')
df_edu.head()
```

```python
ax = df_edu.plot(y=["High school or less", "Some college", "College graduate"], kind='line', color=['r','b','g'])
vals = ax.get_yticks()
ax.set_yticklabels(['{:.0%}'.format(x) for x in vals])
plt.title("% of U.S. adults who use at least one social media site, \n by education level")
plt.legend(loc='best')
plt.savefig('outputs/02_sm_use_by_edu.png')
plt.show();
```
