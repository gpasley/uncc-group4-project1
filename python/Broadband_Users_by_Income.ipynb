```python
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Loading the CSV
broadband_users_df = pd.read_csv("../Resources/Broadband_Users.csv")
broadband_users_df.head()
```

```python
#Setting X Axis and Tick Locations
x_axis = np.arange(len(broadband_users_df))
tick_locations = [value for value in x_axis]
```

```python
#Creating a list indicating where to write X Labels

plt.figure(figsize=(20,4))
plt.bar(x_axis, broadband_users_df["% of U.S. Adults who are home broadband users by Income"], color='r', alpha=1.0, align="center")
plt.xticks(tick_locations, broadband_users_df["Unnamed: 1"], rotation="vertical")
```

```python
# 30,000 to 49,999

plt.figure(figsize=(20,4))
plt.bar(x_axis, broadband_users_df["% of U.S. Adults who are home broadband users by Income"], color='r', alpha=1.0, align="center")
plt.xticks(tick_locations, broadband_users_df["Unnamed: 2"], rotation="vertical")
```
