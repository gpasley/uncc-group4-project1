```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
obscene_material = [5791, 6026, 7292, 8516]
x_axis = np.arange(len(obscene_material))
plt.bar(x_axis, obscene_material, color="black", align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["2011", "2012", "2013", "2014"])
plt.xlabel("Year")
plt.ylabel("Number of Reported Incidents(Per Year)")
plt.title("Reported Pornography/Obscene Material")
plt.tight_layout()
plt.savefig("../Project/Obscene_Material.png")
```
