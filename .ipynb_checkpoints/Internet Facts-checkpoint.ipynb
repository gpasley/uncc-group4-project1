```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# filepath = "Dataset/internet_social_media_facts/social_media_fact_sheet.xlsx"
df = pd.read_excel("Dataset/internet_social_media_facts/social_media_fact_sheet.xlsx", sheet_name=2, skiprows=1)
df['White'].fillna(0, inplace=True)
df['Black'].fillna(0, inplace=True)
df['Hispanic'].fillna(0, inplace=True)
#df.columns = ['date', 'White', 'Black', 'Hispanic']
df = df.rename(columns={'Unnamed: 0':'date'})
df.head()
```

```python
df2 = df.set_index('date')
df2.head()
```

```python
df3 = df2.style.format({
    'White':'{:.2%}',
    'Black':'{:.2%}',
    'Hispanic':'{:.2%}'
})
df3
```

```python
#df3.plot(kind='line', y='White', color='red')
#df3.plot(kind='line', y='Black', color='blue')
#df3.plot(kind='line', y='Hispanic', color='green')
df2.plot(y=["White", "Black", "Hispanic"], kind='line', color=['r','b','g'])
plt.show();
```
